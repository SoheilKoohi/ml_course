```python
import pandas as pd
```

```python
df = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")
```

```python
df_sample = df.sample(frac=1).reset_index(drop=True)
```

```python
df_sample.shape
```

```python
df.shape
```

```python
import uuid
df_sample['id'] = df_sample.apply(lambda _: uuid.uuid4(), axis=1)
```

```python
cols = list(df_sample.columns)
cols = [cols[-1]] + cols[:-1]
df_sample = df_sample[cols]
```

```python
from sklearn.model_selection import train_test_split
train , test = train_test_split(df_sample, test_size=0.2, random_state=1)
```

```python
test[test["Attrition"]=="Yes"].shape[0] / test.shape[0]
```

```python
train[train["Attrition"]=="Yes"].shape[0] / train.shape[0]
```

```python
train.to_csv("hr_train.csv", index=False)
test.to_csv("hr_test.csv", index=False)
```
